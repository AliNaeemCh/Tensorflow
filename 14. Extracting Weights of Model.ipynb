# Extracting Weights of Model

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/HousingData.csv')
```

```python
df_cleaned = df.dropna()
```

```python
X = df_cleaned.drop(columns=['MEDV'])
y = df_cleaned['MEDV']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
mm  = MinMaxScaler()
X_train = mm.fit_transform(X_train)
X_test = mm.fit_transform(X_test)
```

```python

model = tf.keras.Sequential([
    tf.keras.Input(shape= (X_train.shape[1],) ),
    tf.keras.layers.Dense(units=4, activation='relu', kernel_regularizer=tf.keras.regularizers.L2(0.01)),
    tf.keras.layers.Dense(units=1, activation='linear', kernel_regularizer=tf.keras.regularizers.L2(0.01))
])
model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
model.fit(
    X_train,
    y_train,
    batch_size=1000,
    epochs=100)
```

```python
for layerNum, layer in enumerate(model.layers):
    weights = np.array(layer.get_weights()[0])
    biases = layer.get_weights()[1]
    print(f'Layer Number: {layerNum+1}\n')
    for neuron in range(len(biases)):
        print(f'Neuron: {neuron+1} (Layer {layerNum+1})\n')
        print('Weight vector:\n', weights[:, neuron], '\n') 
        print('Bias: ', biases[0], '\n') 
```
